# Using Beautiful soup
First we import the beautifulsoup library

```python
import os
import pandas as pd
from bs4 import BeautifulSoup
```

Next we make the soup, by passing the path to the HTML file into a file handle, then passing that file handler into the BeautifulSoup constructor along with a parser.

```python
#list of dictionaris to build file and later convert to a dataframe
df_list = []
folder = r'C:\Users\user\Downloads\rt-html\rt_html'
for movies_html in os.listdir(folder):    
    with open(os.path.join(folder, movies_html)) as file:
        soup = BeautifulSoup(file, 'lxml')
        title = soup.find('title').contents[0][:-len(' - Rotten Tomatoes')]
        audience_score = soup.find('div',class_='audience-score meter').find('span').contents[0].strip().replace('%','')
        num_audience_ratings = soup.find('div',class_='audience-info hidden-xs superPageFontColor')
        num_audience_ratings = num_audience_ratings.find_all('div')[1].contents[2].strip().replace(',','')
        
        #append to list of dictionaries
        df_list.append({'title': title,
                       'audience_score': int(audience_score),
                       'number_of_audience_ratings': int(num_audience_ratings)})
df = pd.DataFrame(df_list, columns = ['title','audience_score','number_of_audience_ratings'])
df


```
